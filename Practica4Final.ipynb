```python

import numpy as np
import bokeh
import matplotlib.pyplot as plt
import pandas as pd
a1 = [5.0, 10.0, 5.0, 10.0, 5.0, 5.0, 1.0, 1.0, 0.5, 0.0]  #mL de sln KH2PO4 (0.1M)
a2 = [00, 0.5, 1.0, 5.0, 5.0, 10.0, 5.0, 10.0, 10.0, 5.0] #mL de sln Na2HPO4 (0.1M)

acido = 0.240
basico = 0.113 #Tomados a 451nm

acido2 = 0.016
basico2 = 0.954  #TOMADOS A 617nm

absorbancia1 = [1.024, 0.500, 0.652, 0.703, 0.681, 0.493, 0.436, 0.322, 0.343, 0.287] #Datos de absorbancia 1 451nm
absorbancia2 = [0.013, 0.093, 0.285, 0.403, 0.378, 0.553, 0.658, 0.875, 0.861, 0.883] #Datos de absorbancia 2
pH = [4.82, 5.68, 6.30, 6.40, 6.41, 6.85, 7.13, 7.18, 7.30, 7.49]
n = len(absorbancia1)

log = [np.log((absorbancia2[i]*acido)/(absorbancia1[i]*basico2)) for i in range (n)]
#Linear Regression
#Calculo de la pendiente (m) y el intercepto (b) de la linealizacion

#donde almacenaremos los valores de las sumas
suma_xi  = 0
suma_yi  = 0
suma_xixi= 0
suma_yiyi= 0
suma_xiyi= 0
Rss = 0

for i in range( n ):        
    suma_xi   += log[i]
    suma_yi   += pH[i]
    suma_xixi += (log[i]-np.mean(log))**2
    suma_yiyi += (pH[i]-np.mean(pH))**2
    suma_xiyi += (log[i]-np.mean(log))*(pH[i]-np.mean(pH))

#aplicado las formulas

m = suma_xiyi/suma_xixi

b = np.mean(pH)-m*np.mean(log)

r = suma_xiyi/np.sqrt(suma_xixi*suma_yiyi)

# CALCULO DEL RSS, DESVIACION ESTANDAR DE LA REGRESION LINEAL, de m y de b

for i in range( n ):
    Rss += (pH[i]-(m*log[i]+b))**2

VarianzaResidual = Rss/(n-2)

DesvEstReg = np.sqrt(Rss/(n-2))

DesvEstPendiente = DesvEstReg/np.sqrt(suma_xixi)

DesvEstIntercepto = DesvEstReg*np.sqrt(1/n+np.mean(log)**2/suma_xixi)

t = 2.57 #valor de t two tailed n = 5, 95%

#Linea recta aproximada
L = m * np.array(log) + b  #note que aqui toca transformar la lista de x a un arreglo!

#---------------------------------------------------------
print("pendiente", m) #m
print("intercepto", b) #b
print("####################################################")
print("coeficiente de correlacion", r) #r
print("recta", "y = ", str(m), "x +", str(b))
print("Suma de los cuadrados de los residuos: ", Rss)
print("Varianza de los residuos: ", VarianzaResidual)
print("Desviacion estandar de la regresion lineal: ", DesvEstReg)
print("Desviacion estandar de la pendiente: ", DesvEstPendiente)
print("Desviacion estandar del intercepto: ", DesvEstIntercepto)
print("intervalo de confianza de la pendiente", str(m) + " +/- " + str(t*DesvEstPendiente))
print("intervalo de confianza del intercepto", str(b) + " +/- " + str(t*DesvEstIntercepto))

print("####################################################")
```
